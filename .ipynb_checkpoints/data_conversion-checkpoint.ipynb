## Conversion of data and demo

```python
import os
from ase.io import Trajectory
from ase.units import Hartree
import pickle
```

```python
rootdir = 'PETN_training'

for subdir, dirs, files in list(os.walk(rootdir)):
    for file in files:
        
        full_path = str(os.path.join(subdir, file))
        if full_path[-8:] == 'dft.traj':
        
            #make a list of dictionaries
            traj_list = []
            traj_ase = Trajectory(full_path, 'r')
            for atoms in traj_ase:
                data = {'z': atoms.get_atomic_numbers(),
                        'cart': atoms.get_positions(),
                        'E': atoms.get_potential_energy()/Hartree}
                traj_list.append(data)

            #pickle the data in the sub-folders
            with open(os.path.join(subdir, 'trajectory_dft.pkl'), 'wb') as outfile:
                pickle.dump(traj_list, outfile)
```
